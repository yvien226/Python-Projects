```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
```

## Extract tattslotto

```python
# initialise
data = []
num_pages = 16
per_page_ini = 0
freq = 50

# extracting the results on each page
for i in range(1,num_pages):
    list_num = []
    
    print("Extracting on page:"+str(i)+" of 16")
    url = "https://en.lottolyzer.com/history/australia/tattslotto/page/"+str(i)+"/per-page/" + str(per_page_ini) + "/summary-view"
    content = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(content,"html.parser")
    
    for j in soup.find_all('tr'):
        num = j.get_text()
        list_num.append(num)
    
    data.append(list_num[3:])
    
    # update per page
    per_page_ini = per_page_ini + freq
    #print("Done!")
    
print("Done")
```

```python
date = []
winning_no = []
supp_no = []
from_last = []
sum_num = []
avg_num = []
odd_even = []
one_ten = []
ten_twenty = []
twenty_thirty = []
thirty_forty = []
forty_fifty = []

for m in range(len(data)):
    for k in range(len(data[m])):
        result = (data[m][k]).split("\n")[1:len((data[m][k]).split("\n"))-1]
        date.append(result[1])
        winning_no.append(result[2].strip())
        supp_no.append(result[3].strip())
        from_last.append(result[4].strip())
        sum_num.append(result[5].strip())
        avg_num.append(result[6].strip())
        odd_even.append(result[7].strip())
        one_ten.append(result[8].strip())
        ten_twenty.append(result[9].strip())
        twenty_thirty.append(result[10].strip())
        thirty_forty.append(result[11].strip())
        forty_fifty.append(result[12].strip())
        
        d = {'Date':date,'Winning No':winning_no,'Supp No':supp_no,
             'From Last':from_last, 'sum': sum_num, 'average': avg_num,
             'Odd/Even': odd_even, '1-10': one_ten, '11-20': ten_twenty,
             '21-30': twenty_thirty, '31-40': thirty_forty, '41-50': forty_fifty}
        df = pd.DataFrame(data=d)
```

```python
# split odd even, supp number and  winning number
df[['Odd', 'Even']] = df['Odd/Even'].str.split('/', 1, expand=True)
df[['w1', 'w2', 'w3', 'w4', 'w5', 'w6']] = df['Winning No'].str.split(',', 6, expand=True)
df[['s1', 's2']]= df['Supp No'].str.split(',', 2, expand=True)
df = df.drop(['Odd/Even'], axis=1)
```

# Extract powerball

```python
# initialise
data = []
num_pages = 11
per_page_ini = 0
freq = 50

# extracting the results on each page
for i in range(1,num_pages):
    list_num = []
    
    print("Extracting on page:"+str(i)+" of 11")
    url = "https://en.lottolyzer.com/history/australia/powerball/page/"+str(i)+"/per-page/" + str(per_page_ini) + "/summary-view"
    content = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(content,"html.parser")
    
    for j in soup.find_all('tr'):
        num = j.get_text()
        list_num.append(num)
    
    data.append(list_num[3:])
    
    # update per page
    per_page_ini = per_page_ini + freq
    #print("Done!")
    
print("Done")
```

```python
date = []
winning_no = []
powerball = []
from_last = []
sum_num = []
avg_num = []
odd_even = []
one_ten = []
ten_twenty = []
twenty_thirty = []
thirty_forty = []

for m in range(len(data)):
    for k in range(len(data[m])):
        result = (data[m][k]).split("\n")[1:len((data[m][k]).split("\n"))-1]
        date.append(result[1])
        winning_no.append(result[2].strip())
        powerball.append(result[3].strip())
        from_last.append(result[4].strip())
        sum_num.append(result[5].strip())
        avg_num.append(result[6].strip())
        odd_even.append(result[7].strip())
        one_ten.append(result[8].strip())
        ten_twenty.append(result[9].strip())
        twenty_thirty.append(result[10].strip())
        thirty_forty.append(result[11].strip())
        
        d = {'Date':date,'Winning No':winning_no,'Powerball':powerball,
             'From Last':from_last, 'sum': sum_num, 'average': avg_num,
             'Odd/Even': odd_even, '1-10': one_ten, '11-20': ten_twenty,
             '21-30': twenty_thirty, '31-40': thirty_forty}
        df2 = pd.DataFrame(data=d)
```

```python
# split odd even, and  winning number
df2[['Odd', 'Even']] = df2['Odd/Even'].str.split('/', 1, expand=True)
df2[['w1', 'w2', 'w3', 'w4', 'w5', 'w6', 'w7']] = df2['Winning No'].str.split(',', 7, expand=True)
df2 = df2.drop(['Odd/Even'], axis=1)
```
